## Question 1 

*A student is trying to decide between two GPUs. He want to use the GPU for his research to run Deep learning algorithms, so the only thing he is concerned with is speed.*

*He picks a Deep Learning algorithm on a large data set and runs it on both GPUs 15 times, timing each run in hours. Results are given in the below lists GPU1 and GPU2.*

Hint: You can import ttest function from scipy to perform t tests 

Given,

Null Hypothesis : There is no significant difference between data sets

Alternate Hypothesis : There is a significant difference

*Do two-sample testing and check whether to reject Null Hypothesis or not.*

#### Answer:

```python
import pandas as pd
from scipy.stats import ttest_ind,levene,binom
import matplotlib.pyplot as plt


GPU1 = pd.DataFrame([11,9,10,11,10,12,9,11,12,9,11,12,9,10,9])
GPU2 = pd.DataFrame([11,13,10,13,12,9,11,12,12,11,12,12,10,11,13])

ttest,pval = ttest_ind(GPU1,GPU2)
print("p-value",pval,ttest)
#Let level of significance = 0.05
print("p value: ",pval , " < level of significance 0.05 hence reject null hypothesis.There is significant diff between GPU1 and GPU2")

```

# Question 11 - Split data into train, test sets 
Divide the data into training and test sets with 80-20 split using scikit-learn. Print the shapes of training and test feature sets.*

Check: train_test_split function

```python
from sklearn import linear_model
from sklearn.cross_validation import train_test_split


lm = linear_model.LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=0)
lm.fit(x_train, y_train)
```

```python
slope = lm.coef_
print(slope)
```
