```python
import numpy as np
import pandas as pd
```

```python
laliga = pd.read_csv('Laliga.csv', header = 1)

laliga.head(10)
```

```python
#Replacing '-' with 0
laliga = laliga.replace('-',0)
```

```python
laliga.head(10)
```

```python
#Teams that started playing between 1930-1980
teams_debut_between_1930_1980 = laliga[(laliga['Debut'].str[:4].astype(int) >1930) & (laliga['Debut'].str[:4].astype(int) <1980)]
print(teams_debut_between_1930_1980)
```

```python
laliga_points = laliga['Points'].astype(int)
laliga_points = laliga_points.sort_values(ascending  = False).head(5)

top_five_teams_points = laliga.loc[laliga['Points'].astype(int).isin(laliga_points) ]

#Printing top 5 teams in terms of points
top_five_teams_points.loc[:,'Team']
```

```python
#Function that returns all the teams with their goal diff and team which has max and min goal diff
def Goal_diff_count(laliga_data_set):   
    goal_diff = laliga_data_set['GoalsFor'].astype(int) - laliga_data_set['GoalsAgainst'].astype(int)
    laliga_data_set['GoalDifference'] = goal_diff
    laliga_data_set.set_index('Team',inplace = True)
    del laliga_data_set['GoalsFor'], laliga_data_set['GoalsAgainst']
    return laliga_data_set, laliga_data_set.loc[:,'GoalDifference'].idxmax(), laliga_data_set.loc[:,'GoalDifference'].idxmin()



laliga_goal_diff = laliga[['Team','GoalsFor','GoalsAgainst']]

teams_with_goal_diff, team_max_goal_diff, team_min_goal_diff = Goal_diff_count(laliga_goal_diff)

print("All Teams with their goal diff:\n", teams_with_goal_diff)
print("Team with max goal diff: ",team_max_goal_diff)
print("Team with min goal diff: ", team_min_goal_diff)
```

```python
#Calculating winning percentage
winninig_percentage = (laliga['GamesWon'].astype(int)/laliga['GamesPlayed'].astype(int))*100

#Filling Nan with 0
winninig_percentage.fillna(0,inplace = True)

#Appending ne column winning percentage to the dayaset
laliga['WinningPercentage'] = winninig_percentage

winninig_percentage.sort_values(ascending  = False).head(5)
top_five_teams_winning_percentage = laliga.loc[laliga['WinningPercentage'].isin(winninig_percentage)].head(5)

#Printing top 5 teams with winning percentage
print(top_five_teams_winning_percentage)
```

```python
def sum_of_points(points):
    return points.astype(int).sum()

#Grouping teams based on their best position
sum_of_points_by_position = laliga.groupby(['BestPosition'])['Points'].apply(sum_of_points)

#Printing sum of points by grouping teams on best position
print(sum_of_points_by_position)
```
