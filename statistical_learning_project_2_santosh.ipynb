```python
import  pandas as pd
from scipy.stats import ttest_rel,ttest_1samp
```

```python
titan = pd.read_csv('Titan.csv')
titan.head()
```

```python
old_scheme_mean = titan.loc[:,'Old Scheme'].mean()
print("old scheme mean",old_scheme_mean)
new_scheme_mean = titan.loc[:,'New Scheme'].mean()
print("new scheme mean",new_scheme_mean)
```

```python
#h0: old_scheme_mean = new_scheme_mean (null hypothesis)
#h1: new_scheme_mean > old_scheme_mean (alternative hypothesis)
#Since new_scheme_mean is greater than a value(alternative hypothesis) the rejection region lies in the upper side, so a one tail t-test has to be used.
#Let level of significance = 0.05
ttest,pval = ttest_rel(titan.loc[:,'New Scheme'], titan.loc[:,'Old Scheme'])
print("p value:",pval,"t value:",ttest)
#Since we are using one-tail test we use p/2
one_tail_pval = pval/2
print(one_tail_pval)
```

```python
print("Since p value:",one_tail_pval,"is greater than alpha 0.05, we do not reject the null hypothesis. Which means that the new scheme does not have significantly raised outputs")
```
