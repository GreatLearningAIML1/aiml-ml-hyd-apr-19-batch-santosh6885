```
# 1.Read the dataset
import pandas as pd 
df_bank=pd.read_csv('gdrive/My Drive/bank.csv')
```

```
df_bank.head()
```

```
# 2.Drop the columns which are unique for all users like IDs (2.5 points)

# Dropping Rownumber and customer Id
df_bank = df_bank.drop(['RowNumber', 'CustomerId'], axis = 1)

df_bank.columns
```

```
# 3. Distinguish the feature and target set (2.5 points)

# Exited column is the target column and all the remaining columns are features.
X = df_bank.drop('Exited', axis = 1)
Y = df_bank['Exited']
```

```
# Label encoding Gender and geography as they are categorical

from sklearn.preprocessing import LabelEncoder
df_enc = LabelEncoder()
X['Geography'] = df_enc.fit_transform(X['Geography'])
```

```
X['Gender'] = df_enc.fit_transform(X['Gender'])
```

```
# Dropping Surname as it has no contribution to the model
X = X.drop('Surname', axis =1)
```

```
# 4. Divide the data set into Train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```
X_train.columns
```

```
# 5. Normalize the train and test data (2.5 points)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
X_train_transform = scaler.fit_transform(X_train) 
X_test_transform = scaler.fit_transform(X_test) 
```

```
# 6. Initialize & build the model (10 points)

import tensorflow as tf
```

```
import numpy as np
```

```
model_bank = tf.keras.Sequential()
model_bank.add(tf.keras.layers.BatchNormalization())
```

```
model_bank.add(tf.keras.layers.Dense(10,input_shape = (10,), activation='sigmoid'))
model_bank.add(tf.keras.layers.Dense(8, activation='sigmoid'))
model_bank.add(tf.keras.layers.Dense(3, activation='softmax'))
```

```
model_bank.compile(optimizer='sgd', loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```
model_bank.fit(np.array(X_train_transform),np.array(y_train),  epochs=10)
```

```
model_bank.summary()
```

```
# 7. Optimize the model (5 points)

import keras
from keras.models import Sequential
from keras.layers import Dense

model_bank = Sequential()


model_bank.add(Dense(output_dim=10, init='uniform', activation='relu', input_dim=10))

model_bank.add(Dense(output_dim=5, init='uniform', activation='relu'))

model_bank.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))
```

```


model_bank.compile( optimizer='adam', loss='binary_crossentropy', 
                   metrics=['accuracy'])
```

```
model_bank.fit(X_train_transform,y_train,  epochs=100, batch_size=10)
```

```
# 8. predict the results using 0.5 as a threshold (5 points)

y_pred = model_bank.predict(X_test_transform)
```

```
y_pred = (y_pred > 0.5)
```

```
y_pred
```

```
from sklearn.metrics import confusion_matrix
bank_cm = confusion_matrix(y_test, y_pred)
```

```
bank_cm
```

```
model_bank.evaluate(X_test_transform, y_test)
```
