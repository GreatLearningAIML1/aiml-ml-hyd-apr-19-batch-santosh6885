### K - Means Clustering

### 4. Assign a dummy array called distortion

```python
distortion = []
```

You can later use this array to plot the elbow plot

### 5. Calculating distortion - C distance

```python
from sklearn.cluster import KMeans
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

```python
#Initial a list to track SSE for KMeans with different K
sse_df = pd.DataFrame(columns=['K','SSE'])

for i in range(1,11):
    print("*",i)
    
    #Initialize KMeans
    km = KMeans(n_clusters=i)
    
    #Fit on the data
    km.fit(df_attr)
    
    
    #Track K and SSE information in sse_df
    sse_df.loc[sse_df.shape[0]] = [i, km.inertia_]
    
    distortion.append( km.inertia_)
sse_df.head(10)
```

```python
print(distortion)
```

### 6. Plotting Elbow/ Scree Plot

```python
import seaborn as sns

```

Use Matplotlib to plot the scree plot - Note: Scree plot plots distortion vs the no of clusters

```python
sns.lineplot(x = 'K', y = 'SSE', data = sse_df)
```

### Find out the optimal value of K

```python
print("From the elbow plot the optimal value of k can be taken as 4")
```

### Using optimal value of K - Cluster the data. 
Note: Since the data has more than 2 dimension we cannot visualize the data. As an alternative, we can observe the centroids and note how they are distributed across different dimensions

```python
km = KMeans(n_clusters=4, random_state = 10)
km.fit(df_attr)
df_attr['clustelLabels'] = km.predict(df_attr)
df_attr.head(20)
```

You can use kmeans.cluster_centers_ function to pull the centroid information from the instance

```python
km.cluster_centers_
```

### 7. Store the centroids in a dataframe with column names from the original dataset given

```python
df_c = df_attr.drop('clustelLabels', axis = 1)
df_centroids = pd.DataFrame(km.cluster_centers_,columns=df_c.columns)
df_centroids.head()
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
import numpy as np
```

```python
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])

```

```python
d = np.concatenate([a, b,c])
len(d)
```

### 9. Combine all three arrays a,b,c into a dataframe

```python
df_h = pd.DataFrame(d)
df_h.head()

```

### 10. Use scatter matrix to print all the 3 distributions

```python
import seaborn as sns

sns.scatterplot(x = df_h[0],y = df_h[1])
```

### 11. Find out the linkage matrix

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist 
import matplotlib.pyplot as plt 
```

Use ward as linkage metric and distance as Eucledian

```python
Z = linkage(df_h, method = 'ward', metric='euclidean')
c, coph_dists = cophenet(Z , pdist(df_h))
c
```

### 12. Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(15, 15))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z)
plt.tight_layout()
```

### 13. Recreate the dendrogram for last 12 merged clusters 

Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

```python
plt.figure(figsize=(15, 15))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, p =12, truncate_mode = 'lastp')
plt.tight_layout()
```

### 14. From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

```python
# From the above dindogram it can be inferred that to get three clusters the optimal distance is between 25 to 75
```

### 15. Using this distance measure and fcluster function to cluster the data into 3 different groups

```python
f_array = fcluster(Z, t= 30, criterion='distance')

df_h['labels'] = f_array
df_h.head()
```

### Use matplotlib to visually observe the clusters in 2D space 

```python

plt.scatter(x = df_h[0],y = df_h[1], c = df_h['labels'])
```
